```python
import json
import csv  
import re

with open('Lyrics_ЗамайZamay.json', encoding='utf8') as json_file:
    data = json.load(json_file)
    
songs = data['songs']

data_file = open('lyrics.csv', 'w', encoding='utf8', newline ='')
 
# create the csv writer object
csv_writer = csv.writer(data_file)
 
# Counter variable used for writing
# headers to the CSV file
count = 0


csv_writer.writerow(['name', 'lyrics'])

print(songs[0]['lyrics'])


for song in songs:
    print(song['title'])

 
 
    title = re.sub(r'[^\w\s]','',song['title']).replace("\n", " ")
    lyrics = re.sub(r'[^\w\s]','',song['lyrics']).replace("\n", " ")
 
    # Writing data of CSV file
    if lyrics != '':
        csv_writer.writerow([title, lyrics])
 
data_file.close()
```
